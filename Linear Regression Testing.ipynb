```python
import pandas as pd
import RealValueGA
model = RealValueGA.LinearRegression()
df = pd.read_csv(r'lintest.csv')
datapoints = [(i,) for i in df['x'].tolist()]
actual = df['y'].tolist()
new_x, new_y = model.normalize(datapoints, actual)
```

```python
model.run(1, datapoints, actual)
```

```python
import matplotlib.pyplot as plt
plt.scatter(new_x, new_y)
```

```python
import numpy as np
plt.scatter(new_x, new_y)
x = np.linspace(0, 1, 2)
y = 0.7775249050572934*x + 0.051978239963266226
plt.plot(x, y, '-r', label="line of best fit")
plt.show()
```
