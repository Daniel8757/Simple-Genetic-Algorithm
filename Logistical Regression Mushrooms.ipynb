```python
import pandas as pd
from sklearn.utils import shuffle
data = pd.read_csv('mushrooms.csv')
# get y values
yvals = []
for d in data.iterrows():
    if d[1]['class'] == 'p':
        yvals.append(1)
    else:
        yvals.append(0)

# drop x column
data.drop(columns=data.columns[0], axis=1, inplace=True)
```

```python
# fill missing data with mode
for column in data.columns:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

```python
# convert to numerical attributes
dataframes = []
for column in data.columns:
    o_h_encoded = pd.get_dummies(data[column], prefix=str(column))
    dataframes.append(o_h_encoded)
cleaned_data = pd.concat(dataframes, axis=1)
```

```python
# get xvals
xvals = []
for d in cleaned_data.iterrows():
    info = d[1]
    xvals.append(list(info))
```

```python
import RealValueGA
model = RealValueGA.Logis\\\\\\\ticalRegression()
original = sum([(model.run(len(xvals[0]), xvals, yvals, log=False)) for _ in range(4)])/4
```

```python
mut_0 = sum([(model.run(len(xvals[0]), xvals, yvals, mutation_rate=0.0003, log=False)) for _ in range(4)])/4
mut_1 = sum([(model.run(len(xvals[0]), xvals, yvals, mutation_rate=0.003, log=False)) for _ in range(4)])/4
```

```python
import matplotlib.pyplot as plt
plt.xlabel('mutation rate')
plt.ylabel('fitness')
plt.plot([0.0003, 0.001, 0.003], [mut_0, original, mut_1])
```

```python
cross_0 = sum([(model.run(len(xvals[0]), xvals, yvals, crossover_rate=0, log=False)) for _ in range(4)])/4
cross_1 = sum([(model.run(len(xvals[0]), xvals, yvals, crossover_rate=0.3, log=False)) for _ in range(4)])/4
cross_2 = sum([(model.run(len(xvals[0]), xvals, yvals, crossover_rate=1, log=False)) for _ in range(4)])/4
```

```python
plt.xlabel('crossover rate')
plt.ylabel('fitness')
plt.plot([0, 0.3, 0.7, 1], [cross_0, cross_1, original, cross_2])
```

```python
gaus_0 = sum([(model.run(len(xvals[0]), xvals, yvals, gaussian_variance=0.03, log=False)) for _ in range(4)])/4
gaus_1 = sum([(model.run(len(xvals[0]), xvals, yvals, gaussian_variance=0.3, log=False)) for _ in range(4)])/4
```

```python
plt.xlabel('gaussian rate')
plt.ylabel('fitness')
plt.plot([0.03, 0.1, 0.3], [gaus_0, original, gaus_1])
```
