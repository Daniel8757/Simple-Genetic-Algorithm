```python
import pandas as pd
data = pd.read_excel('cc_data.xls')
```

```python
xvals = []
yvals = []
for d in data[1:].iterrows():
    info = d[1]
    xvals.append(info[1:-1])
    yvals.append(info[-1])
```

```python
import RealValueGA
model = RealValueGA.LogisticalRegression()
```

```python
f1 = model.run(len(xvals), xvals, yvals, log=False)
```
